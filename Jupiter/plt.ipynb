# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def linegraph(df_):
    """Return a line graph."""
    df = pd.read_csv(df_)
    df['Date'] = pd.to_datetime(df['Date'])
    #  Date   datetime

    plt.figure(figsize=(12, 7))
    plt.plot(df['Date'], df['Close'], alpha=0.7, label=" ", linewidth=3, color='b')
    plt.title(' ', fontsize=16)
    plt.ylabel(', USD', fontsize=12)
    plt.legend()
    plt.grid()

linegraph("ohlc.csv")
```

```python
def some_linegraph(df_):
    """Return some line graphs."""
    df = pd.read_csv(df_)
    date = df.set_index('Date')
    #   DataFrame    

    date.plot(subplots=True, figsize=(10, 6))
    plt.legend(loc=0)
    plt.suptitle(
        ' ',
        fontsize=16
    )
    plt.xlabel('')
    plt.ylabel(', USD', fontsize=12)

some_linegraph("ohlc.csv")
```

```python
def bargraph(df_):
    """Return a bar graph."""
    df = pd.read_csv(df_)
    plt.figure(figsize=(15, 7))
    plt.bar(df['Date'], df['Profit/Losses'])
    plt.title('    ', fontsize=14)
    plt.xlabel('')
    plt.ylabel(' / , .')

bargraph('budget_data.csv')
```

```python
def piegraph():
    """Return a pie graph."""
    #    
    import numpy as np
    cmap = plt.get_cmap("tab20c")
    outer_colors = cmap(np.arange(3)*4)
    colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

    data = [82, 86, 87, 88, 89, 92, 93, 95]
    keys = ['', ' ', '', '', '', '',
            '', '']

    plt.subplots(figsize=[10, 6])
    plt.pie(x=data, labels=keys, wedgeprops=dict(width=0.2), colors=colors)
    plt.title("  ", fontsize=14)

piegraph()
```
